```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
r = 1.0
t = np.linspace(0, 2 * np.pi, n)
x = r * np.cos(t)
y = r * np.sin(t)
z = np.zeros(n)

d = np.sqrt(x ** 2 + y ** 2 + z ** 2)
c = d / np.max(d)

fig, ax = plt.subplots()
ax.scatter(x, y, c=c, cmap="Reds")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
r1 = 1.0
r2 = 0.5
t = np.linspace(0, 2 * np.pi, n)
x1 = r1 * np.cos(t)
y1 = r1 * np.sin(t)
x2 = r2 * np.cos(t)
y2 = r2 * np.sin(t)
z1 = np.zeros(n)
z2 = np.zeros(n)

d1 = np.sqrt(x1 ** 2 + y1 ** 2 + z1 ** 2)
d2 = np.sqrt(x2 ** 2 + y2 ** 2 + z2 ** 2)
c1 = d1 / np.max(d1)
c2 = d2 / np.max(d2)

fig, ax = plt.subplots()
ax.scatter(x1, y1, c=c1, cmap="Greens")
ax.scatter(x2, y2, c=c2, cmap="Oranges")
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the vertices of the cube
vertices = [[0, 0, 0], [0, 1, 0], [1, 1, 0], [1, 0, 0], [0, 0, 1], [0, 1, 1], [1, 1, 1], [1, 0, 1]]

# Define the edges of the cube
edges = [[0, 1], [1, 2], [2, 3], [3, 0], [4, 5], [5, 6], [6, 7], [7, 4], [0, 4], [1, 5], [2, 6], [3, 7]]

# Calculate the average z value of all vertices
avg_z = sum(vertex[2] for vertex in vertices) / len(vertices)

# Plot the vertices and edges of the cube
for edge in edges:
    x = [vertices[edge[0]][0], vertices[edge[1]][0]]
    y = [vertices[edge[0]][1], vertices[edge[1]][1]]
    z = [vertices[edge[0]][2], vertices[edge[1]][2]]
    
    # Change the color of the edge based on its z value relative to the average z value
    if z[0] > avg_z or z[1] > avg_z:
        color = 'green'
    else:
        color = 'orange'
    ax.plot3D(x, y, z, color)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Create a figure and axis
fig, ax = plt.subplots()

# Set the axis limits
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)

# Initialize a line object to be updated in the animation
line, = ax.plot([], [], 'o-', lw=2)

# Define the initialization function
def init():
    line.set_data([], [])
    return line,

# Define the animation update function
def update(frame):
    x = np.linspace(0, 10, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * frame))
    line.set_data(x, y)
    return line,

# Create the animation using FuncAnimation
ani = FuncAnimation(fig, update, frames=100, init_func=init, blit=True)

plt.show()
```
